### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load dataset

```python
url='C:/Users/ASUS/Desktop/TFS Foundation Tasks/SampleSuperstore.csv'
df= pd.read_csv(url)
print('The first few rows of dataframe:')
print(df.head())
print('******************************************************************')
print('summary statistics of the dataset:')
print(df.describe())
print('******************************************************************')
print('Missing values:')
print(df.isnull().sum())
print('******************************************************************')
print('Duplicate rows:')
print(df.duplicated().sum())


```

### Data visualization

##### Distribution of profit

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Profit'], bins=30, kde=False, color='skyblue')
plt.title('Distribution of Profit')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()
```

##### Profit by Region

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Profit', data=df, ci=None)
plt.title('Profit by Region')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.show()
```

##### Profit by Category

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Category', y='Profit', data=df, ci=None)
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

##### Correlation heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
